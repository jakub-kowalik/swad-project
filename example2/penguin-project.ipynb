```python
import pandas as pd
```

```python
df = pd.read_csv('penguins_size.csv')
df.head()
```

```python
df.isnull().sum()
df = df[df['sex'] != '.']
df.sex.value_counts()
```

```python
df.dropna(subset=['sex'], inplace=True)
df.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
normalize = df.drop(columns=['species', 'island', 'sex'])
mms = MinMaxScaler()
normalize = mms.fit_transform(normalize)
mms.
normalize = pd.DataFrame(normalize)
normalize.columns = mms.get_features_names_out()
normalize
```
